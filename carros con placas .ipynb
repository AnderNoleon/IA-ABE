```python
placa = []
image = cv2.imread('C:/Users/unifo/Documents/auto9.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB));
```

```python
gray = cv2.blur(gray, (3,3))
canny = cv2.Canny(gray, 130, 190)
plt.imshow(canny, cmap = 'gray');
```

```python
gray = cv2.blur(gray, (3,3))
canny = cv2.Canny(gray, 130, 200)
canny = cv2.dilate(canny, None, iterations = 1)
plt.imshow(canny, cmap = 'gray')
```

```python
contorno, _ = cv2.findContours(canny, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
contorno
```

```python
contorno, _= cv2.findContours(canny, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
```

```python
for c in contorno:
    area = cv2.contourArea(c)
    print(area)
    
    
    x, y, w, h = cv2.boundingRect(c)
    epsilon = 0.09 * cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, epsilon, True)
    if len (approx) == 4 and area > 9000:
        print('areaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa ', area)
        print('wwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwww', w)
        print('h ', h)
        aspect_ratio = float(w)/h
        print('relacion de aspecto' , aspect_ratio)
        
        cv2.drawContours(image, [approx], 0 ,(0,255,0),3)
```

```python
for c in contorno:
    area = cv2.contourArea(c)
    print(area)
    
    x, y, w, h = cv2.boundingRect(c)
    epsilon = 0.09 * cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, epsilon, True)
    if len(approx) == 4 and area > 9000:
        print('area ' ,area)
        print('w', w)
        print('h ', h)
        aspect_ratio = float(w)/h
        print('relacion de aspecto' , aspect_ratio)
        
        cv2.drawContours(image, [approx], 0 ,(0,255,0),3)
        if aspect_ratio > 2:
            placa = gray[y: y+h,x:x+w]
            text = pytesseract.image_to_string(placa, config ='--psm 11').strip()
            print('placa', text)
            cv2.rectangle(image, (x,y), (y+w, y+h), (0,255,0),3)
            cv2.putText(image, text , (x-5, y - 5),1,3.0, (0,255,0),3)
```
